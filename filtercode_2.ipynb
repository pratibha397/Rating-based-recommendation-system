```python
import pandas as pd
import numpy as np

# Read JSON file
file_path = r"C:\Users\KIIT\AppData\Local\Temp\Rar$DIa16000.9634\500ECE.json"
df = pd.read_json(file_path, lines=True)

# Filter data based on ASIN and create an independent copy
filter_ids = ['0594451647', '140053271X']
filtered_data_2 = df[df['asin'].isin(filter_ids)].copy()

# Function to compute helpfulness metrics
def compute_helpfulness(helpful):
    if isinstance(helpful, list) and len(helpful) == 2:
        helpful_votes, total_votes = helpful
        if total_votes > 0:
            helpfulness_score = (helpful_votes ** 2) / total_votes  # Formula for b_ij
            return helpful_votes, total_votes, helpfulness_score
    return 0, 0, 0  # Default values if missing or invalid format

# Apply function to create new columns
filtered_data_2[['helpful_votes', 'total_votes', 'helpfulness_score']] = filtered_data_2['helpful'].apply(lambda x: pd.Series(compute_helpfulness(x)))

# **Normalization of helpfulness_score to range [0,1]**
min_score = filtered_data_2['helpfulness_score'].min()
max_score = filtered_data_2['helpfulness_score'].max()
if max_score > min_score:  # Avoid division by zero
    filtered_data_2['helpfulness_score'] = (filtered_data_2['helpfulness_score'] - min_score) / (max_score - min_score)
else:
    filtered_data_2['helpfulness_score'] = 0  # If all values are the same, set to 0

# Compute Most Recent Review Score (most_ij)
n_prime = len(filtered_data_2)
filtered_data_2['z_ij'] = [sum(1 / (e ** 2) for e in range(1, n_prime - i + 1)) for i in range(n_prime)]
filtered_data_2['most_ij'] = filtered_data_2['z_ij'] / filtered_data_2['z_ij'].sum()

# Compute Top Ranking Reviewer Score (top_ij)
filtered_data_2['q_ij'] = (1 / (filtered_data_2['helpful_votes'] + 1)) * (n_prime - filtered_data_2.index)
filtered_data_2['top_ij'] = filtered_data_2['q_ij'] / filtered_data_2['q_ij'].sum()

# Compute Final Reliability Score (d_ij)
alpha = 0.5  # Weighting factor
filtered_data_2['d_ij'] = alpha * filtered_data_2['top_ij'] + (1 - alpha) * filtered_data_2['most_ij']

# Save to CSV
output_path = r"C:\Users\KIIT\Desktop\Project\filtered_data_2.csv"
filtered_data_2.to_csv(output_path, index=False)

# Print the extracted values
print("CSV file saved successfully at:", output_path)
print(filtered_data_2[['asin', 'helpful', 'helpful_votes', 'total_votes', 'helpfulness_score', 'most_ij', 'top_ij', 'd_ij']])
```

```python
import pandas as pd
import numpy as np
from scipy.sparse.linalg import svds

# Load the filtered data with reliability scores
file_path = r"C:\Users\KIIT\Desktop\Project\filtered_data_2.csv"
df = pd.read_csv(file_path)

# Step 1: Create User-Product Reliability Matrix
try:
    user_product_matrix = df.pivot(index="reviewerID", columns="asin", values="d_ij").fillna(0)
    print("\nUser-Product Matrix Shape:", user_product_matrix.shape)
except KeyError as e:
    print(f"Error: Column missing -> {e}")
    exit()

# Step 2: Convert Matrix to NumPy Array
R = user_product_matrix.values
print(R)

# Ensure matrix is not empty
if R.size == 0:
    print("Error: User-Product Matrix is empty!")
    exit()

# Step 3: Perform Singular Value Decomposition (SVD)
k = min(R.shape) - 1 if min(R.shape) > 1 else 1  # Ensure k is valid
U, sigma, Vt = svds(R, k=k)
sigma = np.diag(sigma)

# Step 4: Reconstruct the Matrix
reconstructed_matrix = np.dot(np.dot(U, sigma), Vt)

# Step 5: Extract Top 10 Reliable Users
df_reconstructed = pd.DataFrame(reconstructed_matrix, index=user_product_matrix.index)
top_10_users = df_reconstructed.mean(axis=1).sort_values(ascending=False).head(10)

# Print Top 10 Reliable Users
print("\nTop 10 Most Reliable Users Based on SVD:")
print(top_10_users)
```

```python
import pandas as pd
import numpy as np
from scipy.sparse.linalg import svds
from sklearn.metrics import mean_squared_error

# Load the filtered data with reliability scores
file_path = r"C:\Users\KIIT\Desktop\Project\filtered_data_2.csv"
df = pd.read_csv(file_path)

# Step 1: Create User-Product Reliability Matrix
user_product_matrix = df.pivot(index="reviewerID", columns="asin", values="d_ij").fillna(0)

# Step 2: Convert Matrix to NumPy Array
R_original = user_product_matrix.values

# Step 3: Perform Singular Value Decomposition (SVD)
k = min(R_original.shape) // 2  # Reduce k to avoid perfect reconstruction
U, sigma, Vt = svds(R_original, k=k)
sigma = np.diag(sigma)

# Step 4: Reconstruct the Matrix
R_pred = np.dot(np.dot(U, sigma), Vt)

# Step 5: Compute Loss Function (Frobenius Norm)
loss = np.linalg.norm(R_original - R_pred, ord="fro")  # Frobenius Norm
print("Loss Function (Frobenius Norm):", loss)

# Step 6: Compute Mean Squared Error (MSE) on Non-Zero Entries
non_zero_indices = R_original.nonzero()
mse = mean_squared_error(R_original[non_zero_indices], R_pred[non_zero_indices])

print("Mean Squared Error (MSE):", mse)

# Step 7: Compute Hit Ratio
threshold = 0.5  # Threshold for a "hit"
hits = np.sum((R_pred[non_zero_indices] >= threshold) == (R_original[non_zero_indices] >= threshold))
hit_ratio = hits / len(non_zero_indices[0])

print("Hit Ratio:", hit_ratio)
```
